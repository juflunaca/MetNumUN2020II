```python
import math
x=2
y=0.5

print("For e^(x):")
e_to_2 = x**0/math.factorial(0) + x**1/math.factorial(1) + x**2/math.factorial(2) + x**3/math.factorial(3) + x**4/math.factorial(4)
print(e_to_2)
print(math.exp(x))

print("\nFor cos(x):")
cos2= 1 - x**2/math.factorial(2) + x**4/math.factorial(4) - x**6/math.factorial(6) + x**8/math.factorial(8)
print(cos2)
print(math.cos(x))

print("\nFor sin(x):")
sin2= x - x**3/math.factorial(3) + x**5/math.factorial(5) - x**7/math.factorial(7) + x**9/math.factorial(9)
print(sin2)
print(math.sin(x))

print("\nFor 1/(1-y):")
frac2= 1 + y + y**2 + y**3 + y**4
print(frac2)
print(1/(1-y))

print("\nFor log(1+y):")
log1_05= y - y**2/2 + y**3/3 - y**4/4 + y**5/5  
print(log1_05)
print(math.log(1+y))

print("\nFor log(y):")
log05= (y-1) - (y-1)**2/2 + (y-1)**3/3 - (y-1)**4/4 + (y-1)**5/5
print(log05)
print(math.log(y))
```

```python
print("For e^(x):")
def func_e_to_2(n):
    x = 2
    e_to_2 = 0
    for i in range(n):
        e_to_2 += x**i/math.factorial(i)
    
    return e_to_2

out = func_e_to_2(10)
print(out)

print("\nFor cos(x):")
def func_cos2(n):
    x=2
    cos2 = 0
    for i in range(n):
        cos2 += ((-1)**i)*(x**(2*i))/math.factorial(2*i)
    return cos2
out = func_cos2(10)
print(out)
    
print("\nFor sin(x):")
def func_sin2(n):
    x=2
    sin2 = 0
    for i in range(n):
        sin2 += ((-1)**i)*(x**(2*i+1))/math.factorial(2*i+1)
    return sin2
out = func_sin2(10)
print (out)

print("\nFor 1/(1-y):")
def func_frac2(n):
    y=0.5
    frac2 = 0
    for i in range(n):
        frac2 += y**i
    return frac2
out = func_frac2(10)
print (out)

print("\nFor log(1+y):")
def func_log1_05(n):
    y=0.5
    log1_05 = 0
    for i in range(1,n+1):
        log1_05 += ((-1)**(i+1))*y**i/i
    return log1_05
out = func_log1_05(10)
print (out)

print("\nFor log(y):")
def func_log05(n):
    y=0.5
    log05 = 0
    for i in range(1,1+n):
        log05 += ((-1)**(i+1))*(y-1)**i/i
    return log05
out = func_log05(10)
print (out)
```

```python
print("For e^(x):")
def func_e(x, n):
    e_approx = 0
    for i in range(n):
        e_approx += x**i/math.factorial(i)
    return e_approx
out = func_e(2,10)
print(out)

print("\nFor cos(x):")
def func_cos(x, n):
    cos_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i)
        denom = math.factorial(2*i)
        cos_approx += ( coef ) * ( (num)/(denom) )    
    return cos_approx
out = func_cos(2,10)
print(out)

print("\nFor sin(x):")
def func_sin(x, n):
    sin_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i+1)
        denom = math.factorial(2*i+1)
        sin_approx += ( coef ) * ( (num)/(denom) )    
    return sin_approx
out = func_sin(2,10)
print(out)

print("\nFor 1/(1-x):")
def func_frac(x, n):
    frac_approx = 0
    for i in range(n):
        frac_approx += x**i
    return frac_approx
out = func_frac(0.5,10)
print(out)

print("\nFor log(1+x):")
def func_log_1(x, n):
    log_1_approx = 0
    for i in range(1,n+1):
        coef = (-1)**(i-1)
        log_1_approx += coef * (x**i/i)    
    return log_1_approx
out = func_log_1(0.5,10)
print(out)

print("\nFor log(x):")
def func_log(x, n):
    log_approx = 0
    for i in range(1,n+1):
        coef = (-1)**(i-1)
        log_approx += coef * ((x-1)**i/i)    
    return log_approx
out = func_log(0.5,10)
print(out)
```
