## Se modifica todo el codigo original para la funcion (x-1)(x-2)(x-3)(x-4)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
extent = 0.0, 4.0, 0.0, 4.0 
   
width = 400
height= 400
pixels = np.full((width,height,3),0)
for i in range(width):
  for j in range(height):
    if (j>i+100): 
      pixels[height-j-1,i,0] = pixels[height-j-1,i,1] = pixels[height-j-1,i,2] = 255
      #print(i,j)

plt.title('bisection_convergence')
plt.imshow(pixels,extent = extent)
plt.show()
```

```python
def func(x): 
    return (x-1) *(x-2) *(x-3) * (x-4)


x = np.linspace(0.0,5.0,501)
fig, ax = plt.subplots()
ax.plot(x,func(x))
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.grid(True, which='both')
plt.title('f(x)=(x-1) *(x-2) *(x-3) *(x-4)')
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
width = height = 100
xi=0.0
xf=5.0
eps = 1e-3
mxiter = 1e4

pixels = np.full((width,height,3),0)
# roots
r1 = 1
r2 = 2
r3 = 3
r4 = 4
   
multcol = 5

for i in range(height):
            
            aa =   float(xi + (i * (xf-xi) /height))
            
            for j in range(width):
                bb= float(xi + (j* (xf-xi)/ width))

                a = aa
                b = bb
                c=a
                iter = 0
                
                while ((b-a) >= eps and (func(a) * func(b) <= 0)): 
                   
                    # Check if a is root 
                    if (func(a) == 0.0): 
                      c=a
                      break

                    # Check if b is root 
                    if (func(b) == 0.0): 
                      c=b
                      break
  
                    # Find middle point 
                    c = (a+b)/2
    
                    # Decide the side to repeat the steps 
                    if (func(c)*func(a) < 0): 
                        b = c 
                    else: 
                        a = c 
                    
                    iter = iter + 1
                    
                    if (iter > mxiter):
                     break

                # Pixels colored using the roots
                if abs(c-r1)<eps:
                    # color red
                    pixels[height-j-1,i] = (255 - iter*multcol, 0, 0)
                elif abs(c-r2)<=eps:
                    # color green
                    pixels[height-j-1,i] = (0, 255 - iter*multcol, 0)
                elif abs(c-r3)<=eps:
                    # color blue
                    pixels[height-j-1,i] = (0, 0, 255 - iter*multcol)
                elif abs(c-r4)<=eps:
                    # color magenta
                    pixels[height-j-1,i] = (255 - iter*multcol, 0, 255 - iter*multcol)
                elif (b < a):
                    pixels[height-j-1,i] = (255, 255, 0)
                elif (func(a) * func(b) >= 0):
                    pixels[height-j-1,i] = (255, 255, 255)
                else:   
                    pixels[height-j-1,i] = (0, 0, 0)

                
     
plt.title('bisection_convergence')
plt.xlabel("a in ("+str(xi)+","+str(xf)+")")
plt.ylabel("b in ("+str(xi)+","+str(xf)+")")
extent = xi, xf, xi, xf
plt.imshow(pixels,extent =extent)
plt.show()
```

### Regiones de convergencia:
* Rojo: en esta region la funcion converge hacia x=1, comprende los intervalos (a,b) siempre que a=1, los intervalos (a,b) cuando 0<a<=1 & 1<=b<2, y los intervalos (a,b) cuando 0<a<=1 & 3<b<(4-a).
* Verde: en esta region la funcion converge hacia x=2, comprende los intervalos (a,b) siempre que a=2 y siempre que b=2 excepto para a=1. Tambien comprende los intervalos (a,b) cuando 1<a<=2 & 2<=b<3, los intervalos (a,b) cuando 1<a<=2 & 4<b<(6-a) y los intervalos (a,b) cuando 0<a<1 & b=(4-a).
* Azul: en esta region la funcion converge hacia x=3, comprende los intervalos (a,b) siempre que a=3 y siempre que b=3 excepto para a=1 y a=2. Tambien comprende los intervalos (a,b) cuando 2<a<=3 & 3<=b<4, los intervalos (a,b) cuando 0<a<1 & (4-a)<b<4 y los intervalos (a,b) cuando 1<a<2 & b=(6-a).
* Magenta: en esta region la funcion converge hacia x=4, comprende los intervalos (a,b) siempre que a=4 y siempre que b=4 excepto para a=1, a=2 y a=3. Tambien comprende los intervalos (a,b) cuando 3<a<=4 & 4<=b<5 y los intervalos (a,b) cuando 1<a<2 & (6-a)<b<5.

### Regiones de no convergencia:
* Amarillo: es la region en la que los intervalos no se someten al proceso (y por ende no convergen) ya que a > b, pero se podria interpretar como que se debe realizar el proceso en sentido inverso para el intervalo (b,a), para lo que se esperarian resultados similares reflejados respecto a la linea y=x, es decir en esta region.
* Blanco: es la region en la que la funcion no converge hacia ninguna de las raices puesto que tras hacer muchas interaciones al intervalo original, el intervalo (a,b) resultante tiene un valor del mismo signo para f(a) y f(b), con lo que se puede asumir que no hay una raiz dentro de ese intervalo.
